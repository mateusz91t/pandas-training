### Sticking data frames by `append()`
`append()` can adds only 1 DF to 1 DF

### concat
`concat()` can adds 1 DF to 1 or list of DFs.

```python
supp_and_cust_concatenated = pd.concat(objs=[suppliers, customers])
supp_and_cust_concatenated.iloc[25:35]
```

```python
supp_and_cust_concatenated = pd.concat(
    objs=[suppliers, customers],
    ignore_index=True
)
supp_and_cust_concatenated.iloc[25:35]
```

* `how='left'` : {'left', 'rigth', 'outer', 'inner'}

## `merge()`

args:
* `right` : next DF
* `how='inner'` : {'left', 'right', 'outer', 'inner', 'cross'}
* `on=None` : key to a join from both DF
* ...
* `suffixes=('_x', '_y')` : suffixes that added to columns from left & right side to result
* `indicator=False` : if True it shows differences between left & right, if you type an str, you change the name of the column with the comparison

```python
cat_1 = pd.read_csv(
    './course-files/course-sources/northwind-mongo-master/categories_del_1.csv',
    usecols=['CategoryID', 'CategoryName']
)
cat_1
```

```python
cat_2 = pd.read_csv(
    './course-files/course-sources/northwind-mongo-master/categories_del_2.csv',
    usecols=['CategoryID', 'CategoryName']
)
cat_2
```

```python
cat_1.merge(cat_2)  # inner
```

```python
cat_1.merge(cat_2, on='CategoryID')  # x from left and y from right
```

```python
cat_1.merge(cat_2, on='CategoryID', suffixes=['_1', '_2'])  # preety names for left & right
```

```python
cat_1.merge(cat_2, on='CategoryID', suffixes=['_1', '_2'], how='outer')  # can see differences
```

```python
cat_1.merge(cat_2, suffixes=['_1', '_2'], how='outer')  # full outer merged result
```

```python
# can see differences better with preety name of a comparison col
cat_merged = cat_1.merge(cat_2, on='CategoryID', suffixes=['_1', '_2'], how='outer', indicator='differences')
cat_merged
```

```python
# can see differences better with preety name of a comparison col
cat_1.merge(cat_2, on='CategoryID', suffixes=['_1', '_2'], how='outer', indicator='differences')
```

```python
# filter differences
cat_merged[cat_merged['differences'] != 'both']
```

## Left & Right Merge

```python
products = pd.read_csv(
    './course-files/course-sources/northwind-mongo-master/products.csv',
    usecols=['ProductID', 'ProductName']
)
products.head(2)
```

```python
ordered_products = pd.read_csv(
    './course-files/course-sources/northwind-mongo-master/order-details_del.csv',
    usecols=['OrderID', 'ProductID', 'UnitPrice', 'Quantity']
)
ordered_products.head(2)
```

```python
merged = products.merge(right=ordered_products, on='ProductID', how='left', indicator=True)
merged
```

```python
merged['_merge'].value_counts()
```

```python
merged = products.merge(right=ordered_products, on='ProductID', how='right', indicator=True)
merged
```

```python
merged['_merge'].value_counts()
```

```python
merged = ordered_products.merge(right=products, on='ProductID', how='right', indicator=True)
merged
```

```python
merged['_merge'].value_counts()
```

```python
merged[
    merged._merge == 'right_only'
]
```
