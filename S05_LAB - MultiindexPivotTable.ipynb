```python
df['40K'] = df['40K'].apply(pd.to_timedelta)
df['Half'] = df['Half'].apply(pd.to_timedelta)
df['TotalSeconds'] = df['40K'].apply(lambda x: timedelta.total_seconds(x))
df['HalfSeconds'] = df['Half'].apply(lambda x: timedelta.total_seconds(x))
```

```python
df.pivot_table(
    values='TotalSeconds',
    index='Age',
    columns='M/F'
)
```

## L007

```python
df = pd.read_csv(
    './course-files/course-sources/marathon_results_2016.csv',
    index_col='Bib',
    usecols=['Bib','40K','Half','Pace','Age','M/F','Country','State','City']
)

df['40K'] = df['40K'].apply(pd.to_timedelta)
df['Half'] = df['Half'].apply(pd.to_timedelta)
df['TotalSeconds'] = df['40K'].apply(lambda x: timedelta.total_seconds(x))
df['HalfSeconds'] = df['Half'].apply(lambda x: timedelta.total_seconds(x))

df = df.pivot_table(
    values='TotalSeconds',
    index='Age',
    columns='M/F',
)
df.head()
```

```python
df.reset_index(inplace=True)
df.head()
```

```python
df.melt(id_vars='Age').head()
```

```python
df.melt(
    id_vars='Age',
    value_name='TotalSecends'
).head()
```

```python
df.melt(
    id_vars='Age',
    value_name='TotalSeconds',
    var_name='Sex'
).head()
```

```python
# 7
df = pd.read_csv(
    './course-files/course-sources/marathon_results_2016.csv',
    usecols=['Bib','40K','Half','Pace','Age','M/F','Country','State','City']
)
df['40K'] = df['40K'].apply(pd.to_timedelta)
df['Half'] = df['Half'].apply(pd.to_timedelta)

df['TotalSeconds'] = df['40K'].apply(lambda x: timedelta.total_seconds(x))
df['HalfSeconds'] = df['Half'].apply(lambda x: timedelta.total_seconds(x))

df = df.pivot_table(index="Age",columns="M/F",values=["HalfSeconds","TotalSeconds"])
df.head()
```

```python
df.reset_index(inplace=True)
df.head()
```

```python
df.melt(id_vars='Age').head()
```

```python
df.melt(
    id_vars='Age',
#     value_vars=['HalfSeconds','TotalSeconds','F','M'],  # KeyError
    var_name=['Part', 'Sex'],
    value_name='Elapsed'
).head()
```

```python
df.columns
```

```python
list(df.columns[1:])
```

```python
df.melt(
    id_vars=[('Age', '')],
    value_vars=list(df.columns[1:]),
    var_name=['Part', 'Sex'],
    value_name='Elapsed'
).head()
```
