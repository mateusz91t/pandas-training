## Cleaning data

If some data is missing we may use a few ways:
* update null to average
* deleting each row
* deleting null value from row

Parameters of `frame.dropna()`:
* `how='any'`/`'all'`, default any
* `subset=['cols_name', ...]`, default None
* `axis=0`/`1` or `axis='rows'`/`'columns'`, default 0
* `inplace=True`, default False

```python
mcframe = pd.read_csv('./course-files/course-sources/Mcdonalds.csv')
mcframe.head()
```

```python
mcframe.loc[1, 'Calories'] = np.NaN
mcframe.loc[2, 'TotalFat'] = np.NaN
mcframe.head()
```

```python
mcframe.dropna().head()  # delete each row !!!
```

```python
mcframe.dropna(how='all').head()  # all vals in row are null
```

```python
mcframe.dropna(subset=['TotalFat'], how='any').head()
```

```python
mcframe.dropna(subset=['Calories'], how='any').head()
```

```python
mcframe.dropna(subset=['TotalFat', 'Calories'], how='any').head()
```

### axis [oś] in dropna

In default `axis=0`. It is equal to `axis='rows'`.<br/>
`axis=1` equals to `axis='columns'`

```python
mcframe.dropna(axis='rows').head()  # like default `axis=0`
```

```python
mcframe.dropna(axis='columns').head()  # deleted cols Calories & TotFat
```

### duplicated()
* `keep='first'/'last'/False` :
    * first: first is not duplicate, next is duplicate
    * last: last is not a duplicate
    * False: only not duplicated values

```python
clothes['color'].duplicated(keep='first')
```

```python
clothes['color'].duplicated(keep='last')
```

```python
clothes['color'].duplicated(keep=False)
```

```python
unique_values = clothes['color'].duplicated(keep='first')
clothes[~unique_values]  # not
```
