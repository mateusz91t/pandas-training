```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
bank = pd.read_csv(
    '../../../studia/pjatk/sem4/ADD/DataSets/bank.csv',
    sep=';'
)
```

```python
bank
```

```python
bank.info(memory_usage='deep')
```

```python
bank.nunique()
```

```python
bank['job'] = bank['job'].astype('category')
bank['marital'] = bank['marital'].astype('category')
bank['education'] = bank['education'].astype('category')
bank['contact'] = bank['contact'].astype('category')
bank['month'] = bank['month'].astype('category')
bank['poutcome'] = bank['poutcome'].astype('category')

bank['default'] = bank['default'].replace(('yes', 'no'), (True, False))
bank['housing'] = bank['housing'].replace(('yes', 'no'), (True, False))
bank['loan'] = bank['loan'].replace(('yes', 'no'), (True, False))
bank['y'] = bank['y'].replace(('yes', 'no'), (True, False))
```

```python
# if in a Series are not null it is automatically converted to bool
bank['default'] = bank['default'].astype('bool')
bank['housing'] = bank['housing'].astype('bool')
bank['loan'] = bank['loan'].astype('bool')
bank['y'] = bank['y'].astype('bool')
```
