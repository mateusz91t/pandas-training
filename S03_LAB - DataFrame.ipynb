## L010

```python
fortune = pd.read_csv(
    './course-files/course-sources/Fortune_500_2017.csv',
    usecols=['Rank','Title','Employees','Profits','Assets'],
    index_col='Rank'
)
fortune.head()
```

```python
rank_by_employee = fortune['Employees'].rank(ascending=False)
fortune['RankByEmployee'] = rank_by_employee
fortune.sort_values(by='Employees', ascending=False).head(10)
```

```python
rank_by_profits = fortune['Profits'].rank(ascending=False)
fortune['RankByProfits'] = rank_by_profits

fortune.sort_values(by='Profits', ascending=False).head()
```

```python
is_emp_rank_1st_10 = pd.Series(
    data=[True if i in fortune['RankByEmployee'].sort_values().head(10).keys()
         else False
         for i in fortune.index],
    index=fortune.index
)
is_emp_rank_1st_10.sort_values(ascending=False).head(11)
```

```python
# so much easier: is in 1st 10 ?
is_emp_rank_1st_10 = fortune['RankByEmployee'] <= 10
is_emp_rank_1st_10.sort_values(ascending=False).head(11)
```

```python
is_profit_rank_1st_10 = fortune['RankByProfits'] <= 10
is_profit_rank_1st_10.sort_values(ascending=False).head(11)
```

```python
fortune[is_emp_rank_1st_10 & is_profit_rank_1st_10]
```

```python
is_emp_rank_grater_equal_400 = fortune['RankByEmployee'] >= 400
fortune[is_emp_rank_grater_equal_400 & is_profit_rank_1st_10]
```

```python
fuel = pd.read_csv(
    './course-files/course-sources/fuel.csv',
    usecols=['Vehicle ID','Year','Make','Model','Class','Fuel Type','Combined MPG (FT1)', 'Start Stop Technology'],
#     index_col='Vehicle ID',
    low_memory=False
)
fuel.head()
```

```python
list_of_cars = ['Renault', 'Toyota', 'Ford']
is_in_list_of_cars = fuel['Make'].isin(list_of_cars)
# is_in_list_of_cars[is_in_list_of_cars == True]
fuel[is_in_list_of_cars].head()
```

```python
is_no_start_stop_defined = fuel['Start Stop Technology'].isnull()
is_no_start_stop_defined.loc[16447:16455]
```

```python
fuel[is_no_start_stop_defined].loc[16447:16455]
```

```python
is_start_stop_defined = fuel['Start Stop Technology'].notna()
is_start_stop_defined.loc[16447:16455]
```

```python
fuel[is_start_stop_defined].loc[16447:16455]
```

```python
mpg_between_50_60 = fuel['Combined MPG (FT1)'].between(50, 60)
mpg_between_50_60.loc[16447:16455]
```
