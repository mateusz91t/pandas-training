## L005

```python
fortune = pd.read_csv(
    './course-files/course-sources/Fortune_500_2017.csv',
    usecols=['Rank', 'Title', 'Industry', 'Hqlocation', 'Sector']
)
fortune.head()
```

```python
fortune['Sector'] = fortune['Sector'].str.upper()
fortune.head()
```

```python
industry_contains_comp = fortune['Industry'].str.lower().str.contains('comp')
industry_contains_comp.head()
```

```python
fortune[industry_contains_comp].head()
```

```python
city_state = fortune['Hqlocation'].str.split(',', expand=True, n=1)
city_state.head()
```

```python
fortune[['City', 'State']] = city_state
fortune.head()
```

```python
def build_shortcut(row):
    industry = row['Industry']
    out = ''
    industry_splitted = industry.split(' ')
    for word in industry_splitted:
        out += word[0]
    return out
```

```python
# how to check my func
dict_to_shortcut = dict(Industry='Factory Under Newspaper')
build_shortcut(dict_to_shortcut)
```

```python
industry_short = fortune.apply(build_shortcut, axis=1)
industry_short.head()
```

```python
fortune['IndustryShort'] = industry_short
fortune.head()
```
